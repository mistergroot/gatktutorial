# Notebook 1: Download the Rhesus reference genome

In this notebook you can find the names, source location, and code used to download the reference *Macaca mullata* genome used in our analyses.  

## Assemblies

According to the UCSC genome browser there are several *Macaca mulatta* genome assemblies available (http://hgdownload.soe.ucsc.edu/downloads.html#rhesus). To view these by their NCBI Project numbers we search the Genome ID for Rhesus monkey: **215** (https://www.ncbi.nlm.nih.gov/genome/genomes/215). The full assemblies are all of females, so we will also include the Y chromosome sequenced from a different male, available on NCBI (PRJNA253406). The most recent full assembly (**RheMac8**) was completed by Rogers et al. in 2014 (uploaded in 2015), and includes the Y chromosome sequenced from a male. Full information is below: 

### Full genome assembly (Mmul_8.0.1):
Information: https://www.ncbi.nlm.nih.gov/assembly/GCF_000772875.2/#/def  
Downloads: ftp://ftp.ncbi.nlm.nih.gov/genomes/all/GCF/000/772/875/GCF_000772875.2_Mmul_8.0.1  

### Y-chromosome assembly (Mmul_Y_1.0):
Information: https://www.ncbi.nlm.nih.gov/assembly/GCA_001270425.1  
Downloads: ftp://ftp.ncbi.nlm.nih.gov/genomes/all/GCA/001/270/425/GCA_001270425.1_Mmul_Y_1.0

## Download the data

```python
%%bash

# save ftp download link as a variable
mmul8="ftp://ftp.ncbi.nlm.nih.gov/genomes/all/GCF/000/772/875/GCF_000772875.2_Mmul_8.0.1/GCF_000772875.2_Mmul_8.0.1_genomic.fna.gz"

# make directory for storing reference file
mkdir -p ./reference-genome

# download file to dir
curl -Lk $mmul8 -o ./reference-genome/Mmul8.fna.gz
```
